```python
def connect_db():
    try:
        conn = psycopg2.connect(
            dbname=db_name,
            user=db_user,
            password=db_password,
            host=db_host,
            port=db_port
        )
        return conn
    except Exception as e:
        print(f"Error connecting to the database: {e}")
        return None

def fetch_data(conn, command):
    try:
        df = pd.read_sql(command, conn)
        return df
    except Exception as e:
        print(f"Error executing query: {e}")
        return pd.DataFrame()
```

```python
conn = connect_db()
if not conn:
    print("Failed to connect to the database.")
    exit(-1)
```

## Tarefa: Criação e Alimentação de Tabela na Base `FapCov2103` (Esquema `D2`)

### Estrutura da Tabela

- **Chave identificador**:
  - `Id_Paciente`
  - `Id_Atendimento`
  - `Dt_Coleta`

- **Atributos**:
  - Todos os atributos do paciente
  - `Dt_Atendimento` (data do atendimento)
  - `Dt_Desfecho` (data do desfecho)
  - `De_Desfecho` (valor do desfecho)
  - `De_Exame` (nome do exame)
  - `De_Resultado` (resultado do exame)
  - `Classe` (`CHAR`, calculada):
    - `'P'` → resultado positivo
    - `'N'` → resultado negativo
    - `' '` (em branco) → resultado não identificado

### Critérios de Inclusão

- Somente pacientes que:
  - Possuem desfecho registrado  
  - Possuem exames relacionados a **"Covid"** ou **"Corona"**
