```python
%%sql
UPDATE Configura c
SET Destaques['Distinção'] = sub.nf
FROM (
    SELECT c.nusp, MAX(m.nf) AS nf
    FROM Configura c
        JOIN MatriculaAS m ON c.nusp = m.nusp
    WHERE nf >= 9
    GROUP BY c.nusp
) sub
WHERE c.nusp = sub.nusp
```

```python
%%sql
UPDATE Configura c
SET Destaques['Trancamento'] = sub.sigla
FROM (
    SELECT c.nusp, ANY_VALUE(m.sigla) AS sigla
    FROM Configura c
        JOIN MatriculaAS m ON c.nusp = m.nusp
        WHERE m.frequencia = 0 AND m.nf = 0.5
    GROUP BY c.nusp
) sub
WHERE c.nusp = sub.nusp
```

```python
%%sql
UPDATE Configura c
SET Destaques['Bom'] = sub.qtde_bom
FROM (
    SELECT 
        M.NUSP,
        COUNT(M.nf) AS qtde_bom
    FROM MatriculaAS M
    WHERE M.nf >= 8.0
    GROUP BY M.NUSP
    HAVING COUNT(M.nf) >= 2
) sub
WHERE c.nusp = sub.nusp
```
